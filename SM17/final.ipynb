```python
import matplotlib
import matplotlib.pyplot as plt
# from scipy import *
# from scipy import integrate
# from scipy.integrate import ode
import numpy as np

matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams['font.size'] = 16
```

```python
from sympy.abc import x,y,a,b,c 
from sympy import expand, apart, exp,solve

f = (y-2)*(y+4)

expand(f)

apart(1/f)

f = (y-2)/(y+4)-a*exp(3*x**2)

solve(f,y)

def solution(t,A):
    return((4*A*np.exp(3*t**2)+2)/(1-A*np.exp(3*t**2)))
```

```python
fig = plt.figure(num=1)
ax=fig.add_subplot(111)

x = np.linspace(-2,2,17)
t = np.linspace(-2,2)
y = np.linspace(-6,6,13)
X,Y = np.meshgrid(x,y)

ax.plot(t, solution(t,-0.5))


#Vector field
U = 1
V = X*(Y**2+2*Y-8)
#Normalize arrows
N = np.sqrt(U**2+V**2)  
U2, V2 = U/N, V/N
ax.quiver(X,Y,U2, V2)


plt.xlim([-2,2])
plt.ylim([-6,6])
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.show()
```
