```python
import numpy as np, matplotlib.pyplot as plt
```

```python
def R(x,y):
    return (1.0-x)**2 + (y-x**2)**2
```

```python
plt.figure(figsize=(5,10.5))
plt.axes(aspect='equal')
plt.grid()
plt.xlim(-2.5,2.5)
plt.ylim(-0.5,10)

x = np.linspace(-2,2)
plt.plot(x, R(x,1))

plt.show()
```

```python
plt.figure(figsize=(5,10.5))
plt.axes(aspect='equal')
plt.grid()
plt.xlim(-2.5,2.5)
plt.ylim(-0.5,10)
for section in range(1,6):
    plt.plot(x,R(x,section))
plt.show()
```

```python
plt.figure(figsize=(5,10.5))
plt.axes(aspect='equal')
plt.grid()
plt.xlim(-2.5,2.5)
plt.ylim(-0.5,10)
for section in range(1,6):
    plt.plot(x,R(x,section), label=section)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x,R(x,1),'r-', lw=0.5)
plt.plot(x,R(x,2), 'b--', lw=2)
plt.plot(x,R(x,3), 'go')
plt.plot(x,R(x,4), 'y-.')
plt.show()
```

```python
plt.figure(figsize=(6,4))
y = np.linspace(-2,2)
X,Y = np.meshgrid(x,y)
plt.contour(X,Y,R(X,Y), levels=[1,2,5,10,15,20])
plt.show()
```

```python
plt.figure(figsize=(6,4))
CS =plt.contour(X,Y,R(X,Y), levels=[1,2,5,10,15,20])
plt.clabel(CS, fontsize=9, inline=1)
plt.show()
```
