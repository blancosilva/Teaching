```python
import numpy as np, matplotlib.pyplot as plt
plt.xkcd();
```

```python
pts = np.array([[-1,1],[-1,-1],[1,-1]])

plt.figure()
plt.axes(aspect='equal')
plt.fill(pts[:,0],pts[:,1],'b',alpha=0.2, edgecolor='r')
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.plot([-2,2],[0,0],'k', alpha=0.4)
plt.plot([0,0],[-2,2],'k', alpha=0.4)
plt.title('Feasible region of (P)')
plt.show()
```

```python
x = np.linspace(-2,2)
y = np.linspace(-2,2)
X,Y = np.meshgrid(x,y)

plt.figure()
plt.axes(aspect='equal')
CS = plt.contour(X,Y,X**4+Y**4, colors='k', levels=[0.0625,0.125,0.25,0.5,1,2,3,4], alpha=0.4)
plt.clabel(CS)
plt.fill(pts[:,0],pts[:,1],'b',alpha=0.2, edgecolor='r')
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.plot([-2,2],[0,0],'k',alpha=0.4)
plt.plot([0,0],[-2,2],'k',alpha=0.4)
plt.title('Feasible region of (P)\nSuperimposed level lines')
plt.show()
```

```python
plt.figure()
plt.axes(aspect='equal')
plt.fill(pts[:,0],pts[:,1],'b',alpha=0.2, edgecolor='r')
plt.gca()
plt.xlim(-1.5, 1.1)
plt.ylim(-1.5, 1.1)
plt.plot([-2,2],[0,0],'k', alpha=0.2)
plt.plot([0,0],[-2,2],'k', alpha=0.2)
plt.arrow(0,-0.5,0,-0.55,fc='b',ec='b', head_width=0.05, head_length=0.1, alpha=0.9)
plt.arrow(-1,-1,-np.sqrt(2)/4,-np.sqrt(2)/4,fc='b',ec='b', head_width=0.05, head_length=0.1, alpha=0.9)
plt.plot([-0.35,0.35],[-0.5,-0.5],'k--')
plt.plot([-1.25,-0.75],[-0.75,-1.25],'k--')
cone = [[-1,-1,np.cos(angle),np.sin(angle)] for angle in np.linspace(-np.pi/4+0.2,3*np.pi/4-0.2,8)]
X,Y,U,V = zip(*cone)
plt.quiver(X,Y,U,V, angles='xy', scale_units='xy', scale=3, color='r')
cone = [[0,-0.5,np.cos(angle),np.sin(angle)] for angle in np.linspace(0.2,np.pi-0.2,8)]
X,Y,U,V = zip(*cone)
plt.quiver(X,Y,U,V, angles='xy', scale_units='xy', scale=3, color='r')
plt.title('Improving directions\n(in red)')
plt.show()
```

```python
plt.figure()
plt.axes(aspect='equal')
plt.fill(pts[:,0],pts[:,1],'b',alpha=0.2, edgecolor='r')
plt.gca()
plt.xlim(-1.5, 1.1)
plt.ylim(-1.5, 1.1)
plt.plot([-2,2],[0,0],'k', alpha=0.2)
plt.plot([0,0],[-2,2],'k', alpha=0.2)
plt.arrow(0,0,0.25,0.25,fc='b',ec='b', head_width=0.05, head_length=0.1, alpha=0.9)
plt.arrow(-1,-1,-0.25,0,fc='b',ec='b', head_width=0.05, head_length=0.1, alpha=0.9)
plt.arrow(-1,-1,0,-0.25,fc='b',ec='b', head_width=0.05, head_length=0.1, alpha=0.9)
plt.plot([-0.25,0.25],[0.25,-0.25],'k--')
plt.plot([-1,-1],[-0.65,-1],'k--')
plt.plot([-1,-0.65],[-1,-1],'k--')
cone = [[-1,-1,np.cos(angle),np.sin(angle)] for angle in np.linspace(0.2,np.pi/2-0.2,8)]
X,Y,U,V = zip(*cone)
plt.quiver(X,Y,U,V, angles='xy', scale_units='xy', scale=3, color='r')
cone = [[0,0,np.cos(angle),np.sin(angle)] for angle in np.linspace(3*np.pi/4+0.2,7*np.pi/4-0.2,8)]
X,Y,U,V = zip(*cone)
plt.quiver(X,Y,U,V, angles='xy', scale_units='xy', scale=3, color='r')
plt.title('Invard pointing directions\nfor binding constraints\n(in red)')
plt.show()
```

```python
from matplotlib.patches import Circle, Ellipse
from matplotlib.collections import PatchCollection

fig, ax = plt.subplots()
ax.set_aspect('equal')
patches = [Circle((0,5), np.sqrt(50)),
           Circle((6,0), np.sqrt(37)),
          Ellipse((0,0), 2*np.sqrt(200),2*np.sqrt(200/3))]
p = PatchCollection(patches, alpha=0.4)
ax.add_collection(p)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.plot(7,6,'r.')
plt.title("The feasibility region\n is the intersection\n of these three objects")
plt.show()
```
