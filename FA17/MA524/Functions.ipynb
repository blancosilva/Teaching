```python
import numpy as np, matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.xkcd();
```

```python
# Rosenbrock function
def R(a,b,x,y):
    return (a-x)**2+b*(y-x**2)**2
```

```python
x = np.linspace(-2,2)
y = np.linspace(-1,3)
X,Y = np.meshgrid(x,y)

plt.figure()
CS = plt.contour(X, Y, R(1,2,X,Y), levels=[0.25,0.5,1,2,4,6,8,12,16,20], colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title("Contour plot of $R_{1,1}$")
plt.show()
```

```python
z=R(1,100,X,Y)
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.view_init(60,40)
ax.plot_surface(X, Y, z, linewidth=0.2, antialiased=True)
plt.title('Graph of $R_{1.1}$')

plt.show()
```

```python
# Weierstrass function
def W(a,b,n,x):
    x_ = np.linspace(0,n,n+1)
    aa = a**x_
    bb = b**x_
    return aa.dot(np.cos(bb*np.pi*x))
```

```python
x = np.linspace(-0.5,0.5,1000)

def w(x):
    return W(0.5,7,256,x)

w = np.vectorize(w)

plt.xkcd()
plt.plot(x,w(x))
plt.title('Continuous, nowhere differentiable!')
plt.show()
```

```python
def f(x,y):
    return np.cos(x)*np.cos(y)*np.exp(np.sqrt(x**2+y**2)/4)

x = np.linspace(-2*np.pi,2*np.pi)
y = np.linspace(-2*np.pi,2*np.pi)
X,Y = np.meshgrid(x,y)

plt.figure()
CS = plt.contour(X, Y, f(X,Y), levels=[-4,-2,-1,0,1,2,4], colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title("Contour plot of $f$")
plt.show()
```

```python
def g(x,y):
    return x*y*(x**2-y**2)/(x**2+y**2)

x = np.linspace(-1,1)
y = np.linspace(-1,1)
X,Y = np.meshgrid(x,y)

plt.figure()
CS = plt.contour(X, Y, g(X,Y), colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title("Contour plot of $g$")
plt.show()
```

```python
def h(x,y):
    return y**2-y**4-x**2

plt.figure()
CS = plt.contour(X, Y, h(X,Y), colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title("Contour plot of $h$")
plt.show()
```

```python
def k(x,y):
    return np.exp(-y)*np.cos(x)

x = np.linspace(-2*np.pi,2*np.pi)
y = np.linspace(-2,0)
X,Y = np.meshgrid(x,y)

plt.figure()
CS = plt.contour(X, Y, k(X,Y), colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title("Contour plot of $k$")
plt.show()
```
