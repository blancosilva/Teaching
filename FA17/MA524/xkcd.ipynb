```python
import matplotlib.pyplot as plt, numpy as np
% matplotlib inline
from matplotlib.patches import Circle, Ellipse
from matplotlib.collections import PatchCollection
plt.xkcd();
```

```python
x = np.linspace(-2,2)
plt.figure()
plt.plot(x,-np.ones_like(x),'r',lw=0.4)
plt.plot(x,x**2-1)
plt.xlim(-2,2)
plt.ylim(-3,4)
plt.text(-0.25,-1.5,'(x*,f(x*))')
plt.show()
```

```python
x = np.linspace(-2,2)
plt.figure()
plt.plot(x,-np.ones_like(x),'r',lw=0.4)
plt.plot(x,x**3-1)
plt.text(-0.25,-2.5,'(x*,f(x*))')
plt.show()
```

```python
x = np.linspace(-2.1,2)
y = np.linspace(-2,-0.5)
t = np.linspace(0,1)
patches = PatchCollection([Ellipse((-2,0),0.07,0.2), Ellipse((-0.5,1.9),0.07,0.2)],alpha=100)
patches.set_array(np.array([0,0]))
fig = plt.figure()
ax = fig.add_subplot(111, autoscale_on=True, xlim=(-2.1,0), ylim=(-1,4))
ax.plot(x,((x+2)**2-1)*(x+2))
ax.plot(y,1.875/1.5*(y+2),'b')
ax.plot(-np.ones_like(t),1.3*t,'r--')
ax.annotate('$tf(y) + (1-t)f(x)$',
            xy=(-1, 1.25), xycoords='data',
            xytext=(-80, 60), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-.2"))
ax.annotate('$f(ty+(1-t)x)$',
            xy=(-1, 0), xycoords='data',
            xytext=(50, -10), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-.2"))
ax.text(-2.1,-0.5,'$(x,f(x))$')
ax.text(-0.4,2, '$(y,f(y))$')
ax.set_title('The segment joining two points of \n the graph is always above the graph!')
ax.add_collection(patches)
plt.show()
```

```python
x = np.linspace(-2.1,2)
y = np.linspace(-2,0)
t = np.linspace(0,1)

fig = plt.figure()
ax = fig.add_subplot(111, autoscale_on=True, xlim=(-2.1,0), ylim=(-1,4))
ax.plot(x,((x+2)**2-1)*(x+2))
ax.plot(y,2*(y+1),'b')
ax.annotate('$(x, f(x))$',
            xy=(-1, 0), xycoords='data',
            xytext=(-50, 50), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-.2"))
ax.set_title('The tangent line to any point of \n the graph is always below the graph!')
plt.show()
```

```python
def f(x):
    return np.log(x**4-2*x**2+2)
```

```python
x=np.linspace(-3,3)
plt.figure()
plt.plot(x,f(x))
plt.title(r'$f(x) = \log(x^4-2x^2+2)$')
plt.show()
```
